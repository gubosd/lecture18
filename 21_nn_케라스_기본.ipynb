### 케라스 맛보기 (회귀)
- 출력층 활성화함수 : 없음
- 비용함수 : MSE

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([1,2,3])
y = np.array([3,5,9])
```

```python
model = Sequential()
model.add(Dense(1, input_shape=(1,), activation='linear')) # 'linear' is default
```

```python
model.compile(loss='mse', optimizer='sgd')
model.fit(x.reshape(-1,1), y, epochs=1000)
```

```python
plt.plot(model.history.history['loss'][:100])
```

```python
model.predict(x.reshape(-1,1))
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x.reshape(-1,1)).ravel(), 'r--')
plt.axis('equal')
```

```python
ws = model.get_weights()
ws
```

```python
w = ws[0][0,0]
b = ws[1][0]
w, b
```

### 케라스 맛보기 (이진분류)
- 출력층 활성화함수 : sigmoid
- 비용함수 : binary_crossentropy
> $$ cost = -\frac{1}{N} \cdot \sum_{i=0}^{N-1} [y_i \log(\hat{y_i}) + (1-y_i) \log(1-\hat{y_i})] $$

<img src="https://donarthurzhang.files.wordpress.com/2017/08/screen-shot-2017-08-11-at-6-41-57-pm.png?w=1024" /><br>
(출처: https://projectsig.com/2017/08/11/logistic-regression-with-a-neural-network-cat-recognizer/)

```python
X = np.array([[0,0],[0,2],[2,0]])
y = np.array([0,1,1])

plt.scatter(X[:,0], X[:,1], c=y, s=100)
plt.axis('equal')
```

```python
model = Sequential()
model.add(Dense(1, input_shape=(2,), activation='sigmoid'))
```

```python
from tensorflow.keras.optimizers import SGD # lr=0.01
```

```python
model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.1), metrics=['acc'])
history = model.fit(X, y, epochs=1000)
```

```python
pred_y = model.predict(X)
pred_y
```

```python
((pred_y.ravel()>0.5)==y).mean()
```

```python
plt.plot(history.history['loss'])
```

```python
ws = model.get_weights()
ws
```

```python
w1 = ws[0][0,0]
w2 = ws[0][1,0]
b = ws[1][0]
w1,w2,b
```

```python
plt.scatter(X[:,0], X[:,1], c=y, s=100)
rng = np.array([-1,2])
plt.plot(rng, -(w1*rng+b)/w2)
plt.axis('equal')
```

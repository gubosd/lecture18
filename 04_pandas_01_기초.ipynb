### 함수 적용
- apply() : 데이터프레임에 칼럼별로 적용
- applymap() : 데이터프레임에 항목별로 적용
- map() : 시리즈에 항목별로 적용

```python
d = pd.DataFrame(np.arange(12).reshape(4,3), columns=list('abc'))
d
```

```python
d.applymap(lambda x: x%2) # 항목 하나하나 계산
```

```python
d.apply(lambda col: col.max() - col.min()) # 칼럼이 인자가 된다 (col => number)
```

```python
d.apply(lambda s: s-s.min()) # col => series
```

```python
d.apply(lambda s: (s-s.mean())/s.std()) # col => series (normalization)
```

```python
d.apply(lambda s: pd.Series([s.max()-s.min(), s.mean()]))
```

```python
d.apply(lambda s: pd.Series([s.max()-s.min(), s.mean()], index=['diff','mean']))
```

```python
d.apply(lambda s: pd.Series([s.sum(), s.mean(), s.std()], index=['sum','mean','std']))
```

```python
d.describe()
```

```python
d.applymap(lambda x: x%3)
```

```python
d.a.map(lambda x: x**2) # Series 에만 적용 가능
```

```python
d.a.apply(lambda x: x**2)
```

```python
d.a.apply(lambda x: pd.Series([x,x**2,x**3], index=['x1','x2','x3']))
```

```python
d.a
```

```python
d.a.map({0:'zero', 1:'one', 3:'three', 6:'six'}) # 코드화
```

- concat() : 붙이기

```python
pd.concat([a,b], ignore_index=True)
```

```python
c = pd.DataFrame([[200,201,202]])
pd.concat([a,c])
```

```python
a = pd.DataFrame(np.arange(9).reshape(3,3), columns=list('abc'))
b = pd.DataFrame(np.arange(100,109).reshape(3,3), columns=list('xyz'))
c = pd.concat([a,b], axis=1)
c
```

```python
a = pd.DataFrame([['hong',22,'seoul','hong@naver.com'],
                  ['jang',33,'incheon','jang@gmail.com'],
                  ['sin',61,'kangwon',''],
                  ['dooley',11,'etc','dooley@andromeda.uni']],
                columns=['name','age','address','email'])
a
```

```python
b = pd.DataFrame([['seoul',999], ['busan',700], ['incheon',503], ['etc',300]],
                 columns=['location', 'pop'])
b = b.set_index('location')
b
```

```python
c = pd.DataFrame([['seoul',999], ['busan',700], ['incheon',503], ['etc',300]],
                 columns=['location', 'pop'])
c
```

```python
help(pd.merge)
```

```python
pd.merge(a,b, left_on='address', right_index=True) # how='inner'
# select a.*, b.pop from a, b where a.address=b.location
```

```python
pd.merge(a,c, left_on='address', right_on='location')
```

```python
pd.merge(a,b, left_on='address', right_index=True, how='outer')
# select a.*, b.pop from a outer join b where a.address=b.location
```
